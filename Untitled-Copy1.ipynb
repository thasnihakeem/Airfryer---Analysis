```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 7-day time series dataset
df = pd.read_csv('final.csv')

# Get the unique list of product URLs from the dataset with attributes
product_urls = pd.read_csv('unique.csv')['urls'].unique()

# Create an empty dictionary to store the number of out-of-stock products by date
out_of_stock_dict = {}

# Loop through each day in the 7-day dataset
for date in df['date'].unique():
    # Get the URLs of the products that are in stock on the current day
    in_stock_urls = df[df['date'] == date]['urls'].unique()
    
    # Find the URLs of the products that are out of stock on the current day
    out_of_stock_urls = set(product_urls) - set(in_stock_urls)
    
    # Store the number of out-of-stock products for the current day in the dictionary
    out_of_stock_dict[date] = len(out_of_stock_urls)

# Convert the dictionary to a Pandas DataFrame
out_of_stock_df = pd.DataFrame.from_dict(out_of_stock_dict, orient='index', columns=['out_of_stock'])

# Reset the index of the DataFrame to make it easier to plot
out_of_stock_df = out_of_stock_df.reset_index().rename(columns={'index': 'date'})

# Create a line plot of the number of out-of-stock products over time
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(data=out_of_stock_df, x='date', y='out_of_stock', color=sns.color_palette("rocket")[2], ax=ax,linewidth=2, marker='o')
ax.grid(True)
# Set the title and axes labels
ax.set_title('Out-of-Stock Products Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Number of Out-of-Stock Products')

# Add count labels on the plot
for x, y in zip(out_of_stock_df['date'], out_of_stock_df['out_of_stock']):
    plt.text(x, y, str(y), ha='center', va='bottom', fontweight='bold')

plt.show()

```

```python
for i in range(len(df)):
    if df['capacity'].iloc[i]in ['Not Available', 'default']:
        df['capacity'].iloc[i]='Not Available'  

def convert_to_litres(value):
    value_str = str(value)
    if 'litres' in value_str:
        return round(float(value_str.split(' ')[0]), 1)
    elif 'Quarts' in value_str:
        return round(float(value_str.split(' ')[0]) * 0.946353, 1)
    elif 'Cubic Centimetres' in value_str:
        return round(float(value_str.split(' ')[0]) / 1000, 1)

    
df['capacity'] = df['capacity'].apply(convert_to_litres)  


df['capacity_range'] = ''

for i in range(len(df)):
    if df['capacity'].iloc[i] == 'Not Available':
        df['capacity_range'].iloc[i] = 'Not Available'
    elif df['capacity'].iloc[i] == 'Small':
        df['capacity_range'].iloc[i] = 'Others'
    elif df['capacity'].iloc[i] == 'Large':
        df['capacity_range'].iloc[i] = 'Others'
    elif df['capacity'].iloc[i] is None:  # add this condition
        df['capacity_range'].iloc[i] = 'Not Available'
    else:
        rating = float(df['capacity'].iloc[i])
        if rating < 5.0:
            df['capacity_range'].iloc[i] = '1_5.0'
        elif 5.0<= rating < 10.0:
            df['capacity_range'].iloc[i] = '5.0_10.0'
        elif 10.0<= rating <15.0:
            df['capacity_range'].iloc[i] = '10.0_15.0'    
        elif 15.0<= rating <20.0:
            df['capacity_range'].iloc[i] = '15.0_20.0'
        elif 20.0<= rating < 25.0:
            df['capacity_range'].iloc[i] = '20.0_25.0'
        elif rating >=25.0:
            df['capacity_range'].iloc[i] = '25.0_and_above'
```

```python
def weight_to_grams(value):
    value_str = str(value)
    if 'Kilograms' in value_str:
        return int(round(float(value_str.split(' ')[0]) * 1000))
    elif 'Grams' in value_str:
        return int(round(float(value_str.split(' ')[0])))
    elif 'Pounds' in value_str:
        return int(round(float(value_str.split(' ')[0]) * 453.592))
    elif 'kg' in value_str:
        return int(round(float(value_str.split(' ')[0]) * 1000))
    elif 'g' in value_str:
        return int(round(float(value_str.split(' ')[0])))
    else:
        return value
```

```python
df['Weight'] = df['Weight'].apply(weight_to_grams)
```

```python
df['Weight_range'] = ''
for i in range(len(df)):
    if df['Weight'].iloc[i] == 'Not Available':
        df['Weight_range'].iloc[i] = 'Not Available'
    if df['Weight'].iloc[i] == 'not available':
        df['Weight_range'].iloc[i] = 'Not Available'
    else:
        rating = int(df['Weight'].iloc[i])
        if rating < 1000:
            df['Weight_range'].iloc[i] = '1_1000'
        elif 1000<= rating < 5000:
            df['Weight_range'].iloc[i] = '1000_5000'
        elif 5000<= rating <10000:
            df['Weight_range'].iloc[i] = '5000_10000'    
        elif rating >=10000:
            df['Weight_range'].iloc[i] = '1500_and_above'

df['Weight_range'].value_counts()
```
